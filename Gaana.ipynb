```python
del df['time']
```

```python
lst = []
for i in df['duration']:
  if len(i.split(':')) == 3:
    hour = int(i.split(':')[0])
    mins = int(i.split(':')[1])
    secs = int(i.split(':')[2])
    duration = hour*3600+mins*60+secs
    lst.append(duration)

  elif len(i.split(':')) == 2:
    mins = int(i.split(':')[0])
    secs = int(i.split(':')[1])
    duration = mins*60 + secs
    lst.append(duration)
```

```python
df['duration_in_secs'] = lst
```
